```python
def pnl (text, prices, trades):
  # names = ["Valentina", "Vinnie", "Vladimir", "Vivian", "Celeste", "Colin", "Carlos", "Camilla", "Pablo", "Penelope", "Percy", "Petunia", "Ruby", "Remy", "Rihanna", "Raj", "Amelia", "Adam", "Alina", "Amir"]
  names = trades["buyer"].unique()
  print("- for " + f"{text}")
  for name in names:
      temp1 = trades[trades["buyer"] == f"{name}"]
      temp2 = trades[trades["seller"] == f"{name}"]
      buys = (temp1["price"]*temp1["quantity"]).sum()
      sells = (temp2["price"]*temp2["quantity"]).sum()
      position = temp1["quantity"].sum()-temp2["quantity"].sum()
      mid = int(prices[-1:]["mid_price"].iloc[0])
      print(f"{name} pnl: " + str(sells-buys+position*mid) + ", position: " + str(position))
pnl("strawberries1", prices_strawberries1, df1_strawberries)
pnl("strawberries2", prices_strawberries2, df2_strawberries)
pnl("strawberries3", prices_strawberries3, df3_strawberries)
```

```python
pnl("roses1", prices_roses1, df1_roses)
pnl("roses2", prices_roses2, df2_roses)
pnl("roses3", prices_roses3, df3_roses)
```

```python
pnl("chocolate1", prices_chocolate1, df1_chocolate)
pnl("chocolate2", prices_chocolate2, df2_chocolate)
pnl("chocolate3", prices_chocolate3, df3_chocolate)
```

```python
pnl("basket1", prices_basket1, df1_basket)
pnl("basket2", prices_basket2, df2_basket)
pnl("basket3", prices_basket3, df3_basket)
```
